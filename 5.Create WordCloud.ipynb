```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = 'Here I am coming..'

cloud = WordCloud(background_color='white').generate(text)

plt.imshow(cloud)
plt.axis('off')
plt.show()
```

## Wordcloud different shapes

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image                         # import pillow library
import numpy as np                            # used to convert image into array

text = 'B2 information, especially facts or numbers, collected to be examined and considered and used to help decision-making, or information in an electronic form that can be stored and used by a computer: The data was/were collected by various researchers'

our_mask = np.array(Image.open('cloud_mask.png'))   # download image and load it in same folder, then here converting into array

cloud = WordCloud(background_color='white', mask=our_mask).generate(text)     # use 'mask' parameter

plt.imshow(cloud)
plt.axis('off')
plt.show()
```
